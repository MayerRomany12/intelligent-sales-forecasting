# Milestone 1: Data Collection, Exploration, and Preprocessing

In this notebook, we will explore the Olist dataset and prepare it for forecasting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
sys.path.append(os.path.abspath('../'))

from src.data_preprocessing import load_and_merge_datasets, clean_data, aggregate_time_series

sns.set_theme(style="whitegrid")
```

## 1. Load and Merge Data

```python
raw_data_dir = '../data/raw'
merged_df = load_and_merge_datasets(raw_data_dir)
merged_df.head(20)
```

## 2. Data Cleaning

```python
cleaned_df = clean_data(merged_df)
cleaned_df.info()
```

## 3. Exploratory Data Analysis (EDA)
Let's look at the top selling product categories.

```python
top_categories = cleaned_df.groupby('product_category_name_english')['price'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 Product Categories by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Category')
plt.show()
```

## 4. Aggregate to Time Series
We aggregate the data into a daily frequency to forecast daily sales.

```python
ts_df = aggregate_time_series(cleaned_df, freq='D')
ts_df.head()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(ts_df.index, ts_df['total_sales'], label='Daily Sales', color='blue')
plt.title('Daily Total Sales (Revenue) Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales (BRL)')
plt.legend()
plt.show()
```

```python
# Save the processed data for Milestone 2
ts_df.to_csv('../data/processed/time_series_sales.csv')
print("Saved time_series_sales.csv")
```
